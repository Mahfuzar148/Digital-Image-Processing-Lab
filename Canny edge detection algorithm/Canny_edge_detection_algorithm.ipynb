```python

from google.colab import files
import cv2
import matplotlib.pyplot as plt


uploaded = files.upload()

img = cv2.imread("image.png")

```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

blur = cv2.GaussianBlur(gray, (5,5), 1.4)

# ---- 5️⃣ Canny edge detection ----
edges = cv2.Canny(blur, 100, 200)

# ---- 6️⃣ Show all results ----
plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); plt.title("Original"); plt.axis("off")
plt.subplot(1,3,2); plt.imshow(gray, cmap="gray"); plt.title("Gray"); plt.axis("off")
plt.subplot(1,3,3); plt.imshow(edges, cmap="gray"); plt.title("Canny Edges"); plt.axis("off")
plt.tight_layout();
plt.show()
```
